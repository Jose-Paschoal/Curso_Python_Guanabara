## Desafio 003 - Soma de valores

```python
valor1 = int(input('digite um valor: '))
valor2 = int(input('digite outro valor: '))
soma = (valor1 + valor2)
print('A soma do valor {} mais o valor {} é igual a {}'.format(valor1, valor2, soma))
```

## Desafio 004 - Dissecando uma variável

```python
a = input('Digite algo: ')
print('O tipo primitivo desse valor é', type(a))
print('Só tem espaços?', a.isspace())
print('É um número?', a.isnumeric())
print('É alfabético?', a.isalpha())
print('É alfanumérico?', a.isalnum())
print('Está em maiúsculas?', a.isupper())
print('Está em em minúsculas?', a.islower())
print('Está capitalizada?', a.istitle())
```

## Desafio 005 - Antecessor e Sucessor

```python
n = int(input('Digite um número: '))
print('Para o número digitado {} seu atencessor é {} e seu sucessor é {}'.format(n, (n-1), (n+1)))
```

## Desafio 006 - Dobro, Triplo e Raiz Quadrada

```python
n = int(input('Digite um número: '))
print('O dobro de {} é {}'.format(n, 2*n))
print('O triplo de {} é {}'.format(n, 3*n))
print('A raiz quadrada de {} é {:.2f}'.format(n,n**(1/2) ))
```
